```python
# 조건에 따라 angle_zone 컬럼 생성
def categorize_angle(theta):
    if theta < 68:
        return '1'
    elif 68 <= theta < 90:
        return '2'
    elif 90 <= theta < 112:
        return '3'
    else:
        return '4'

df['angle_zone'] = df['theta_p'].apply(categorize_angle)
```

```python
# 조건에 따라 distance_zone 컬럼 생성
def categorize_distance(distance):
    if distance < 85:
        return 'A'
    elif 85 <= distance < 125:
        return 'B'
    elif 125 <= distance < 170:
        return 'C'
    elif 170 <= distance < 210:
        return 'D'
    elif 210 <= distance < 255:
        return 'E'
    elif 255 <= distance < 295:
        return 'F'
    else:
        return 'G'

df['distance_zone'] = df['distance'].apply(categorize_distance)
```

```python
# zone 컬럼 생성 (angle_zone과 distance_zone 결합)
df['zone'] = df['distance_zone'] + df['angle_zone']
```

```python
# 새로운 매핑 기준
zone_mapping = {
    'A1': 1, 'A2': 2, 'A3': 2, 'A4': 3,
    'B1': 4, 'B2': 5, 'B3': 5, 'B4': 6,
    'C1': 7, 'C2': 8, 'C3': 8, 'C4': 9,
    'D1': 10, 'D2': 11, 'D3': 11, 'D4': 12,
    'E1': 13, 'E2': 14, 'E3': 15, 'E4': 16,
    'F1': 17, 'F2': 18, 'F3': 19, 'F4': 20,
    'G1': 21, 'G2': 22, 'G3': 23, 'G4': 24
}

# 매핑 적용하여 새로운 zone 값 저장
df['zone'] = df['zone'].map(zone_mapping)
```
